MATPLOTLIB

data visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
###Draw graph: a line in a diagram from one positon(say A) to some other positon (say B)
xpoint=np.array([1,2])
ypoint=np.array([5,10])
#here A(1,5) & B(2,10)
plt.plot(xpoint,ypoint)
plt.show()
```

```python
xpoints=np.array([1,6])
ypoint=np.array([4,5])
plt.plot(xpoints,ypoint,'o')
 #here we gave a marker symbol so it will plot only the posisition points and not the line
plt.show()
```

```python
points=np.array([3,8,1,10])
plt.plot(points,'*''g') 
plt.show()
#Note that is only one array is mention under plots then By default taken as values for Y-Axis
#note that comma is not used b/w marker symbol and marker color
#BY DEFAULT the values x taken as 0,1,2,3,4..
```

```python
ypoints=np.array([3,8,1,12])
xpoints=np.array([5,6,7,9])
plt.plot(xpoints,ypoints,marker='X',ms=15,color='red')
plt.show()
#ms= marker size
```

```python
points=np.array([3,8,1,10])
plt.plot(points,marker='o',ms=15,mec='k',color='y')
#mec: marker EDGE color
plt.show()
```

```python
plt.plot(points,marker='*',ms=20,mec='red',mfc='pink')
#not that color is not mentioned(BY DEFAULT THE COLOR IS BLUE)
plt.show()
#mfc=marker Face color
#used both mec and mfc arguments to color the entire marker
```

```python
###Line formating
plt.plot(points,linestyle='dashed',linewidth=3,marker='o',mec='hotpink',ms=15,color='blue',mfc='y')
plt.show()
```

```python
#the shorter form of keyword argument linewidth that can be used is lw to chang width of line.
plt.plot(points,lw=5)
plt.show()
#by DEFAULT the color is blue (why?? bcoz...because it was the color closest to black in its contrast to white, thus making it the next most legible color choice for web pages.)
```

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,5,11])

plt.plot(y1)
plt.plot(y2)
#two line plots in one graph with 1st default color= blue and 2nd as orange
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y,color="pink")
plt.xlabel("Average Pulse",fontsize=12,color='k',fontstyle='italic')
plt.ylabel("Calorie Burnage",fontsize=12,color='k',fontstyle='oblique')
plt.title("Sports Watch data",fontsize=20,color='grey')

plt.grid(color='y',linestyle='dotted')

plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,10,15,20,25])

plt.scatter(x,y)
```

-Bar and horizoontal bar graphs

```python
x=np.array(['a','b','c','d'])
y=np.array([3,8,1,10])

plt.bar(x,y,color="hotpink",width=0.2)
plt.show()
```

```python
#bars displayes horizontaly it is like a and y axis their positions interchanged
x=np.array(['a','b','c','d'])
y=np.array([3,8,1,10])

plt.barh(x,y,color="cyan")
plt.show()
```

- Histograms

```python
p=np.random.normal(100,5,2500)
#normal(loc=0.0, scale=1.0, size=None)  ###these are the by default values
plt.hist(p)
```

-Pie chart

```python
y=np.array([10,135,25,15,19])
plt.pie(y)
plt.show()
```

```python
#TO GET LABELS
y=np.array([10,25,15,19])
mylabels=["cherries","apples","bananas","oranges"]

plt.pie(y,labels=mylabels)
plt.show()
#note that the by default color order is : blue(cherries),orange(apples),green(bananas),red(oranges)
```

```python
#giving a strat angle(note the rotation starts from blue then orange,..)
plt.pie(y,labels=mylabels,startangle=180)
##By default the start angle is 0
plt.show()
```

```python
#seperating the slices from the center
myexplode=[0,0.95,0.8,0.2]
#0:blue, 0.95:orange, 0.8:green, 0.2:red
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
#giving users colors
mycolor=['brown','orange','red','green']
plt.pie(y,labels=mylabels,colors=mycolor)
plt.show()
```

```python
#labels can be represented as legends

plt.pie(y)
plt.legend(mylabels)
plt.show()
```
